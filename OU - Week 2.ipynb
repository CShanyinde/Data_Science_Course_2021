```python
from pandas import *
```

```python
data = read_excel('WHO POP TB some.xls')
data
```

```python
#Display the deaths column
deaths=data['TB deaths']
deaths
```

```python
#Display the population column
pop=data['Population (1000s)']
pop
```

```python
sum=deaths.sum()
sum
```

```python
min=deaths.min()
min
```

```python
max=deaths.max()
max
```

```python
mean=deaths.mean()
mean
```

Note - The median is the middle number in a sorted, ascending or descending, list of numbers and can be more descriptive of that data set than the average. The median is sometimes used as opposed to the mean when there are outliers in the sequence that might skew the average of the values

```python
median=deaths.median()
median
```

```python
mean/median
```
